```python
import numpy as np
inflation_rates = {
    2017: 0.0210,  # 2.10% inflation rate for 2017
    2018: 0.0190,  # 1.90% inflation rate for 2018
    2019: 0.0230   # 2.30% inflation rate for 2019
}
def adjust_salary(salary, from_year, to_year):
    # Ensure inflation_rates is populated correctly
    inflation_rate = 1.0
    for year in range(from_year, to_year):
        if year in inflation_rates:
            inflation_rate *= (1 + inflation_rates[year])
        else:
            raise ValueError(f"Missing inflation rate for year {year}")
    # Calculate adjusted salary
    adjusted_salary = salary * inflation_rate
    return adjusted_salary

def adjust_salary_row(row):
    survey_year = row['Survey Year']
    salary = row['SalaryUSD']

    # Check for NaN values
    if pd.isna(survey_year) or pd.isna(salary):
        return np.nan

    # Ensure correct data types
    try:
        survey_year = int(survey_year)
        salary = float(salary)
    except ValueError:
        return np.nan

    return round(adjust_salary(salary, survey_year, 2019), 0)


df['Survey Year'] = pd.to_numeric(df['Survey Year'], errors='coerce')
df['SalaryUSD'] = pd.to_numeric(df['SalaryUSD'], errors='coerce')
# Apply inflation adjustment to SalaryUSD column
df['AdjustedSalaryUSD'] = df.apply(lambda row: adjust_salary_row(row), axis=1)

# Print updated DataFrame
df
```

```python
df = df.drop(columns=['YearsWithThisDatabase'])
```

```python
def combine_databases(primary, other):
    if pd.isna(other):
        return primary
    primary_set = set(primary.split(', '))
    other_set = set(other.split(', '))
    combined_set = primary_set.union(other_set - primary_set)
    return ', '.join(combined_set)

# Apply the function to each row
df['CombinedDatabases'] = df.apply(lambda row: combine_databases(row['PrimaryDatabase'], row['OtherDatabases']), axis=1)
df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb
from sklearn.preprocessing import StandardScaler

# Assuming df is your original DataFrame and final_df is a copy
final_df['adjustedsalaryusd'] = df['adjustedsalaryusd']

# Calculate the correlation matrix
correlation_matrix = final_df.corr()

# Extract the correlations with the target variable 'adjustedsalaryusd'
target_correlation = correlation_matrix['adjustedsalaryusd'].drop('adjustedsalaryusd')

# Get the top 5 correlated features
top_5_features = target_correlation.abs().sort_values(ascending=False).head(5).index

# Prepare the data
X = final_df[top_5_features]
y = final_df['adjustedsalaryusd']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the XGBoost model
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=200,
    learning_rate=0.01,
    max_depth=4,
    min_child_weight=1,
    subsample=0.6,
    colsample_bytree=0.6,
    gamma=0,
    reg_lambda=1.5,
    reg_alpha=0,
    random_state=42
)

xgb_model.fit(X_train, y_train)

# Make predictions
y_pred = xgb_model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"R-squared: {r2}")
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Optionally, you can also print the feature importances
print("Feature Importances:", xgb_model.feature_importances_)


```
